Quick recap: x_i = P * X_i where (x_i, X_i) is the ith correspondence, and P is a 3x4 dimensional matrix. P is further decomposed as P = K * [ R | T].


```python
#Helper functions
#appends a ones column to input
def hom_coords(x,X):
    n = x.shape[0]
    size = (n,1)
    ones = np.ones(size)
    x = np.concatenate(
        (x,ones),
        axis = 1)
    X = np.concatenate(
        (X,ones),
        axis = 1)
    return x, X

x,X = hom_coords(x,X)
print(x.shape)
print(X.shape)
```

#### Explanation  

The Direct Linear Transform (DLT) algorithm is used to map any object point(X) to an image point(x) using a Projection matrix(P).
$$x = PX$$
$$x = K[R|T]X$$ 
$$x = KR[I_3|-X_0]X$$ 

where $K$ is the camera caliberation matrix , $R$ is the rotation matrix and $X_0$ is the projection center (translation $T$) 

* In this problem we assume that the camera is uncalibrated,i.e, we don't know the K matrix. We have 11 unknowns(6 intrinsic and 5 extrinsic parameters).
* Each world point $(X,Y,Z)$ gives two observation equations for the image coordinates($x$ and $y$).

$$\left[\begin{array}{c}x\\ y\\ 1\end{array}\right] = P \left[\begin{array}{c}X\\ Y\\ Z\\ 1 \end{array}\right]$$
$$x = \frac{p_{11} X + p_{12}Y + p_{13}Z + p_{14}}{p_{31} X + p_{32}Y + p_{33}Z + p_{34}}$$

$$x = \frac{p_{21} X + p_{22}Y + p_{23}Z + p_{24}}{p_{31} X + p_{32}Y + p_{33}Z + p_{34}}$$ 
    Therefore, we need at least 6 points to solve DLT


* We thus pick at least 6 points(x) on the cube image and then estimate the corresponding world points(X) taking the **origin of the world frame (0,0,0) as the left end of the cube** and considering each side as 0.04 units.
* We first rearrange the DLT equation: $$x_i = PX_i = \left[\begin{array}{c}p_{11} & p_{12} & p_{13} & p_{14}\\ p_{21} & p_{22} & p_{23} & p_{24}\\ p_{31} & p_{32} & p_{33} & p_{34} \end{array}\right]X_i = \left[\begin{array}{c}A^T \\ B^T \\ C^T \end{array}\right] X_i $$ 

$$ -X_i^TA + x_iX_i^TC = 0 $$ $$X_i^TB + y_iX_i^TC = 0 $$
* To estimate the elements of $P$, we rearrange the equations and write them as:
$$a_{x_i}^T p = 0$$
$$a_{y_i}^T p = 0$$

    where $$a_{x_i} = [-X_i,-Y_i,-Z_i,-1,0,0,0,0,x_i*X_i,x_i*Y_i,x_i*Z_i,x_i]$$
    $$a_{y_i} = [0,0,0,0,-X_i,-Y_i,-Z_i,-1,y_i*X_i,y_i*Y_i,y_i*Z_i,y_i]$$
* Appending all $a_{x_i}$ and $a_{y_i}$ together for all the points we get a matrix $M$ which is $2I$x$12$ and multiplying with $p$ which is 12x1, we should try to minimise the value of $Mp$
* We apply SVD on the matrix $M$, $$M = USV^T$$ $P$ matrix is the last column of the $V$ matrix and then reshaped to 4x3.
* To get the $K$, $R$ and $T$ matrix, $$P = KR[I_3|-X_0] = [H|h]$$  $$H = KR$$  $$h = -KRX_0$$  
* QR Decomposition of $H$ is done for getting rotation($R$) and calibration matrix($K$) 
$$H = KR $$
* Then projection centre($X_0$) is found by: $$H^{-1} = R^TK^{-1}$$ $$X_0 = -H^{-1}h$$ 

* Then we reproject the points with the P matrix, check the error with MSE and plot the new image points found onto the image of cube.


```python

#DLT function takes as input 
#the image points and world points
#return a 3x4 matrix
def DLT(image_pt,world_pt):
    
    M = []
    no_pts = 6
    for i in range(no_pts):
        X,Y,Z = world_pt[i,0],world_pt[i,1],world_pt[i,2]
        x,y = image_pt[i,0],image_pt[i,1]
        a_x = [-X,-Y,-Z,-1,0,0,0,0,x*X,x*Y,x*Z,x]
        a_y = [0,0,0,0,-X,-Y,-Z,-1,y*X,y*Y,y*Z,y]
        M.append(a_x)
        M.append(a_y)
    M = np.array(M)
    M = M.reshape(2*no_pts,12)
    
    U,S,VT = np.linalg.svd(M)
    L = VT[-1,:] /VT[-1,-1]
    
    #Camera projection matrix:
    P = L.reshape(3,4)
    print("Projection Matrix")
    print(P)    
    
    return P

P = DLT(x,X)
```

## Analysing P

The next part of this question is to analyse how good our estimate is. Report the reprojection error of your P matrix

```python
#Get reprojected points
def get_points(P,X):
    points = P@X.T
    points = points / points[2]
    points = points.T[:,0:2]
    return points

#Get reprojection error
def get_err(x, X, P):
    points = get_points(P, X)
    """
    find the MSE between reprojected points
    and x
    """
    err = 0

    pts = []
    for i in range(6):
        pt = x[i][:2]/x[i][-1]
        pts.append(pt)
    pts = np.array(pts)

    diff = np.subtract(points,pts)
    sq = np.square(diff)
    mean = sq.mean()

    err = mean
    
    #to err is to human
    return err

Reproj_error = get_err(x, X, P)
print("Reprojection_error of P matrix : {}".format(Reproj_error))
```

Given the P matrix, we can estimate K, R and T. Write a function that returns K,R,T given P.
Note: 
P = K * [R | T] where K is a 3x3 matrix and [R | T], P are 3x4 matrices

```python
def ret_RKT(P):
    
    H = P[0:3,0:3]
    h = P[:,-1]
    
    H_inv = np.linalg.pinv(H)
    
    #projection center X0= -H_inv.h
    T = -1*np.matmul(H_inv,h)
    
    #QR Decomposition of H_inv for getting rotation and calibration matrix 
    q, r = np.linalg.qr(H_inv)
    
    K = np.linalg.pinv(r)
    K = r/r[-1,-1]
    
    
    R = q.transpose()

    return R, K , T 

R, K , T = ret_RKT(P)
print("Rotation matrix:")
print(R)
print(" ")
print("Camera Calibration matrix K :")
print(K)
print(" ")
print("Projection_center T")
print(T)
print(" ")
```

Draw epipolar lines on the first image for the corresponding points in the second image.

```python
data_file = "q2/data.txt"

with open(data_file, 'r') as f:
    A = f.readlines()
    
img_points1 = A[2:12]
img_points1 = [i.strip()[:-1] for i in img_points1]
img_points1[0] = img_points1[0][1:]
img_points2 = A[15:25]
img_points2 = [i.strip()[:-1] for i in img_points2]
img_points2[0] = img_points2[0][1:]

F_mat = A[28:31]
F_mat = [i.strip()[:-1] for i in F_mat]
F_mat[0] = F_mat[0][1:]

img_1 = []
img_2 = []
F_MAT = []

for i in img_points1:
    img_1.append(np.fromstring(i[1:-1], sep=' '))
    
for i in img_points2:
    img_2.append(np.fromstring(i[1:-1], sep=' '))
    
for i in F_mat:
    F_MAT.append(np.fromstring(i[1:-1], sep=' '))

img_1 = np.array(img_1)
img_2 = np.array(img_2)
F_mat = np.array(F_MAT)
```

Provide a clear explanation on your approach for both this task and the next:

* Epipolar points (epipoles) are the respective projections of the camera centres on the opposite image plane.  They are the intersections of the baseline with the image planes. <br>
* If the image planes are converging, only then we can have epipoles. Or else, Baseline intersects the image plane at infinity and therefore, Epipoles are at infinity and Epipolar lines are parallel to x axis. <br>



<br>
* Let I1 and I2 be the two image planes.
  e1 and e2 are epipoles. e1 is the image of o2 in I1 and e2 is the image of o1 in I2. <br>
* We note here that, All the world points that map to x1 in I1 (that is the pre-image of x1) map to a line l2 in I2. This line l2 is an epipolar line. That is, the corresponding points lie on conjugate epipolar lines. <br>

* Basically, e1 gives the location of camera_1 as seen by the camera_2 and e2 gives the location of camera_2 as seen by the camera_1 <br>

* Similarly, <br>
* All the world points that map to x2 in I2 (that is the pre-image of x2) map to a line l1 in I1. This line l1 is an epipolar line. The plane containing this line l1 is an epipolar plane (similarly, The plane containing the line l2 is also an epipolar plane) <br>

* The E and F matrices (3x3) encode the epipolar geometry of two views. In the case of a generic world and different cameras, we can obtain the essential and fundamental matrices as follows, <br>

* E depends only on the extrinsic parameters
<br>
* Obtaining E and F matrices: <br>
* From the equation of a straight line, we can write that, <br>

$$ \lambda_1 x_1 = X $$ <br>
$$ \lambda_2 x_2 = RX + T = R \lambda_1 x_1 + T $$

$\hat{T}$ repersents the cross-product operation. That is, Cross product of any two vectors: AxB can be written as the matrix product $\hat{A}B$

$$\hat{A} = \left[\begin{array}{ccc} 0 & -A_z & A_y \\ A_z & 0 & -A_x \\ A_y & A_x & 0 \end{array}\right]$$

$$ \hat{T} \lambda_2 x_2 = \hat{T} R \lambda_1 x_1 + 0 = R \lambda_1 x_1 + T$$ <br>
$$ \lambda_2 x_2^{T} \hat{T} x_2 = \lambda_1 x_2^{T} \hat{T} R x_1 $$ <br>
$$ x_2^{T} \hat{T} R x_1 = 0 $$ <br>
$$ x_2^{T} E x_1 = 0 $$ <br>
$$ \: \:or\: \: x_1^{T} E x_2 = 0 $$ <br>

<br>
E is the Essential Matrix <br>
Similaryly, we get the Fundamental matrix, that is <br>
$$ x_1 = K_1 X $$ <br>
$$ x_2 = K_2 X $$ <br>
$$ x_2^{T} K_2^{-1} \hat{T} R K_1^{-1} x_1 $$ <br>
$$ x_1^{T} F x_2 = 0 $$ <br>

<br>
F is the Fundamental Matrix <br>
The following hold for both E and F

<br>
Now, we see that <br>
$$ l_1 = F^{T} x_2 $$ <br>
$$ l_2 = F x_1 $$ <br>

<br>
Here, l1 and l2 are [a1,b1,c1] and [a2,b2,c2] parameters of the two epipolar lines respectively.
<br>

Now, we are given a point in image_1, and we need to determine the corresponding epipolar line in the image_2, search along that line and find the point corresponding to the point in image_1. <br>

From the above proof, We see that we need to multiply the point in image_1 with F_mat to obtain the epipolar line (to be searched in the image_2, if we want to get the epipolar point). <br>

```python
#Write code to plot epipolar lines on the images
im1 = cv2.imread('./q2/img1.jpg')
im2 = cv2.imread('./q2/img2.jpg')

# function to plot the img_1 and img_2 array points on the images
def display_points(flag, img1, img_1):
    #axes = plt.gca()
    plt.imshow(img1)
#     plt.hold(True)
    plt.scatter(img_1[:,0], img_1[:,1])
#     plt.hold(False)
    plt.savefig('img_given_points' + str(flag) + '.png')
    plt.show()

# function to plot the epipolar lines    
def abline(img1, flag, lines1, slopes, intercepts):
    
    axes = plt.gca()
    plt.imshow(img1)
#     plt.hold(True)
    for i in range(len(slopes)):
        x_vals = np.array(axes.get_xlim())
        y_vals = intercepts[i] + slopes[i] * x_vals
        axes.plot(x_vals, y_vals)
        #plt.xlim((-3000, 3000))
        #plt.ylim((-3000, 3000))
#         axes.hold(True)
    
#     plt.hold(False)
    plt.savefig('op_img' + str(flag) + '.png')
    plt.show()
    
lines1 = []
slopes = []
intercepts = []

for i in range(len(img_1)):
    img = np.append(img_1[i], 1)
    
    # obtaining the line, as described in the theory part of the solution
    ep_line_1 = np.dot(F_mat, img)
    slope = -ep_line_1[0]/ep_line_1[1]
    intercept = -ep_line_1[2]/ep_line_1[1]
    lines1.append(ep_line_1)
    slopes.append(slope)
    intercepts.append(intercept)
    
flag = 1
display_points(flag, im1, img_1)
abline(im1, flag, lines1, slopes, intercepts)
print("a,b,c coefficients of the epipolar lines\n")
print(lines1)

lines2 = []
slopes2 = []
intercepts2 = []

for i in range(len(img_2)):
    img = np.append(img_2[i], 1)
    
    # obtaining the line, as described in the theory part of the solution
    ep_line_2 = np.dot(np.transpose(F_mat), img)
    slope = -ep_line_2[0]/ep_line_2[1]
    intercept = -ep_line_2[2]/ep_line_2[1]
    lines2.append(ep_line_2)
    slopes2.append(slope)
    intercepts2.append(intercept)

flag = 2
display_points(flag, im2, img_2)
abline(im2, flag, lines2, slopes2, intercepts2)
print("a,b,c coefficients of the epipolar lines\n")
print(lines2)
```
